```python
import json
```

```python
def load_data(filename):
    with open(filename,'r') as f:
        data=json.load(f)

    return data
```

```python
data=load_data("store_data.txt")
print(data)
```

```python
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    
    for user in data:
        #Clean ratings
        raw_rating=str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating=text_to_num[raw_rating]    
        user["rating"]=raw_rating 

        #handling missing values
        raw_age=user.get("age")
        if raw_age==None:
            user["age"]=None

        #deduplication
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
    
```

```python
data=clean_data(data)
```

```python
#get insights

def get_insights(data):
    total_rating = 0
    count=0
    for user in data:
        total_rating+=float(user["rating"])
        if(float(user["rating"])<3):
            count+=1
    print(f"Avg rating is :",total_rating/len(data))

    
    print("%age of poor rating is :",(count/len(data))*100)
```

```python
get_insights(data)
```
